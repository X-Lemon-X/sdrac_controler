```python
from sympy import symbols, Matrix, sin, cos, pi, Symbol, eye, acos ,sqrt, asin


def Rz(theta,x,y):
  R = Matrix([[cos(theta), -sin(theta), x],
              [sin(theta), cos(theta), y],
              [0, 0, 1]])
  return R


q1 = Symbol('q1')
q2 = Symbol('q2')
q3 = Symbol('q3')
l1 = Symbol('l1')
l2 = Symbol('l2')
l3 = Symbol('l3')
a = symbols('a')

a01 = Rz(q1,0,0)
a12= Rz(a,q2,l1)
a23= Rz(q3,l2,0)
a34 = Rz(0,l3,0)
display(a01)
display(a12)
display(a23)
display(a34)

A02 = a01*a12
A03 = A02*a23
A04 = A03*a34
display(A02)
A03.simplify()
display("A03")
display(A03)

A04.simplify()
display(A04)
# display(A04_simplified)


q1 = pi/4
q2 = 0
q3 = (210 * pi / 180) + (60 * pi / 180)
l1 = 0.6
l2 = 0.85
l3 = 0.3
a = 0

A04_evaluated = A04.subs(
  {Symbol('q1'): q1, 
  Symbol('q2'): q2, 
  Symbol('q3'): q3,
  Symbol('l1'): l1,
  Symbol('l2'): l2,
  Symbol('l3'): l3,
  Symbol('a'): a
  })


display(A04_evaluated)
display(A04_evaluated.evalf())



```
